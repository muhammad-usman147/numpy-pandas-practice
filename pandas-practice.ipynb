```python
import pandas as pd
import numpy as np
```

```python
x= lambda a: 'Male' if a%2==0 else 'Female'
np.random.seed(5)

dic={
    'name':['usman khan','Quratulain','Raheem','Kareema khan niazi','Huzaifa','Anila','Qazi','Qareema','Omam khan','Oaima','Saif','ALice'],
    'age':np.random.randint(10,30,size=[12]),
    'Gender':[x(i) for i in range(0,12)]
}
df=pd.DataFrame(dic)
df
```

```python
def show(x):
    if x<18:
        return "Child"
    else:
        return 'Adult'
    
df['Maturity']=df['age'].map(show)
```

```python
df
```

```python
def show(x):
    if x=='Male':
        return "Handsome"
    else:
        return 'Good '
    
df['Personality']=df['Gender'].map(show)
```
