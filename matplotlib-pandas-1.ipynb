```python
#adding labels
s=np.array([1,2,3,4,5])
plt.ylim(0,36)
plt.xlim(-1,6)
plt.xlabel('These are the values of x axis') #this show the x axis label
plt.ylabel('These are the values of y axis') #this shows the y axis label
plt.plot(s,s**2)
```

```python
#adding title
s=np.array([1,2,4,6,8])
plt.ylim(0,100)
plt.xlim(-1,10)
plt.xlabel('These are the values of x axis') 
plt.ylabel('These are the values of y axis') 
plt.title("This is a learning plot ") #adding title of this graph

plt.plot(s,s**2)
```

```python
#adding labels
s=np.array([1,2,4,6,8])
plt.ylim(0,100)
plt.xlim(-1,10)
plt.xlabel('These are the values of x axis') 
plt.ylabel('These are the values of y axis') 
plt.title("This is a learning plot ") #adding title of this graph
plt.plot(s,s+2 , label="Line no 1") #this is label 1
plt.plot(s,s**1.5 , label="Line no 2") #this is label 2
plt.plot(s,s**2 , label='Line no 3') #this os label 3 
plt.legend()  #to show the label, you have to use plt.legend
#the box inside the graph of line info is called lagend box
plt.grid()
plt.show()
```

```python
#adding labels in legend method
#adding labels
s=np.array([1,2,4,6,8])
plt.ylim(0,100)
plt.xlim(-1,10)
plt.xlabel('These are the values of x axis') 
plt.ylabel('These are the values of y axis') 
plt.title("This is a learning plot ") #adding title of this graph
plt.plot(s,s+2 ) 
plt.plot(s,s**1.5)
plt.plot(s,s**2 )  
plt.legend(['normal','fast','slow'])  #adding labels in the legend method, sequence wise for the plot()
plt.grid()
plt.show()
```

```python
#adding location of legend box
s=np.array([1,2,4,6,8])
plt.ylim(0,100)
plt.xlim(-1,10)
plt.xlabel('These are the values of x axis') 
plt.ylabel('These are the values of y axis') 
plt.title("This is a learning plot ") 
plt.plot(s,s+2 ) 
plt.plot(s,s**1.5)
plt.plot(s,s**2 )  
plt.legend(['normal','fast','slow'],loc='center') #adding the location of the legend box using loc parameter
plt.grid()
plt.show()
```

```python
#saving the plot
s=np.array([1,3,5,7,9])
plt.ylim(0,100)
plt.xlim(-1,10)
plt.xlabel('These are the values of x axis') 
plt.ylabel('These are the values of y axis') 
plt.title("This is a learning plot ") 
plt.plot(s,s+2 ) 
plt.plot(s,s**1.5)
plt.plot(s,s**2 )  
plt.legend(['normal','fast','slow'],loc='upper left') 
plt.grid()
plt.savefig('grapg.png')
```
