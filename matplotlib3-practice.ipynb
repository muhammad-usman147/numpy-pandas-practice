# Decorate Graphs with Plot  Styles and Types 

```python
x=np.arange(1,5)
plt.plot(x,x*5,label="normal")
plt.plot(x,x*1.5,label="fast")
plt.plot(x,x*2.5,label="slow")
plt.grid()
plt.title("Sample Growth of Measure")
plt.xlabel("Samples")
plt.ylabel("Values Measure")
plt.legend(loc='upper left')
plt.show()
```

```python
#defining colors
x=np.arange(1,5)
plt.plot(x,x*5,'^',label="normal",color='y')
plt.plot(x,x*1.5,'o',label="fast",color='black')
plt.plot(x,x*2.5,"*",label="slow",color='m')
plt.grid()
plt.title("Sample Growth of Measure")
plt.xlabel("Samples")
plt.ylabel("Values Measure")
plt.legend(loc='upper left')
plt.show()
```

```python
#defining some width size style color of marker
y=np.array([1,3,5,4])
plt.plot(y,[2,1,8,1],color='b',marker='o',linestyle='dashdot',linewidth=4,markeredgecolor='black',markerfacecolor='red',markeredgewidth=3,markersize=12)
plt.show()
```

```python
#handling x and y ticks
#ticks defines the range,it shows the ticks where is therre any edge in the graph
x=[5,3,7,2,4,1]
plt.plot(x)
plt.xticks(range(len(x)),['a','b','c','d','e','f'])
plt.yticks(range(1,8,2))
plt.show()
```

```python
#histogram charts
y=np.random.randn(7)
print(y*10)
plt.plot(y*10)
plt.xticks(range(len(y)),['a','b','c','d','e','f','g'])
plt.show()
```

```python
#histogram charts
y=[1,2,3,4,5,6,7,9]
x=[1,4,3,16,25,1,5,1]
t=['cat','dog','apple','hen','lion','coconut','laptop','prisoners']
plt.bar(y,x,tick_label=t)
plt.show()
```

```python
#error bar
x=np.arange(0,5,0.2)
y=np.exp(-x)
e1=np.abs(np.random.randn(len(x)))
e2=np.abs(np.random.randn(len(y)))*0.1
plt.errorbar(x,y,xerr=e2,fmt='^',ecolor='red',capsize=0) #fmt is the plot format for lines,if none, then only
#error bars are plotted with no line connecting them
plt.show()
```

```python
#error bar
x=np.arange(0,5,0.2)
y=np.exp(-x)
e1=np.abs(np.random.randn(len(x)))
e2=np.abs(np.random.randn(len(y)))*0.1
plt.errorbar(x,y,yerr=e1,fmt='^',ecolor='red',capsize=0) #ecolor is the error line color
plt.show() 
```

```python
x=np.arange(0,10)
y=np.arange(10,20)
e1=np.abs(np.random.randn(len(x)))
e2=np.abs(np.random.randn(len(y)))*0.5
plt.errorbar(x,y,yerr=[e2,e1],fmt='.-')
```

```python
#BAR CHARTS, they are displayed in both vertical and horizontal
#bar chart is commonly used for comparision of two or more values
print(x,y)
plt.bar(x,y)
plt.show()
```

```python
dist={'A':10,'B':20,'C':30}
for i,key in enumerate(dist):
    plt.bar(i,dist[key]) 
plt.xticks(np.arange(len(dist))+0.4,dist.keys())
plt.xlim(0,12)
plt.ylim(0.40)
```

```python
np.random.seed(5)
data1=(np.random.rand(5))*10
data2=(np.random.rand(5))*10
data3=(np.random.rand(5))*10
e2=0.5*np.abs(np.random.randn(len(data2)))
locs=np.arange(1,len(data2)+1)
width=0.27
plt.bar(locs,data1,width=width)
plt.bar(locs,data2,yerr=e2,width=width,color='red')
plt.bar(locs+0.2*width,data3,width=width,color='green')
print(e2)
```

```python
#using barh function
np.random.seed(5)
data1=(np.random.rand(5))*10
data2=(np.random.rand(5))*10
data3=(np.random.rand(5))*10
e2=0.5*np.abs(np.random.randn(len(data2)))
locs=np.arange(1,len(data2)+1)
width=0.27
plt.barh(locs,data1)
plt.barh(locs,data2,yerr=e2,color='red')
plt.barh(locs+0.2*width,data3,color='green')
print(e2)
```

```python
#pie chart represents in circular form, where the distributions are called 'wedges'
x=[45,45,30]
label=['CATS','DOGS','CAMEL']
plt.figure(figsize=(5,5))
plt.pie(x,labels=label)
plt.show()
```

```python
#sum attirbutes of pie chart
x=[2,9,21,18,55,20,39]
labels=['Australiv','Britain','India','Ireland','Pakistan','Oman','UAE']
explode=[0.1,0.2,0.3,0,0,0,0]
plt.pie(x,labels=labels,explode=explode) #explode sperates the distributions 
plt.show()
```

```python
#scatter plots, it is use to identify ptential difference between two variable
#and it is also draw before working on a fitting regression function
x=np.random.randn(1000)
y=np.random.randn(1000)
size=50*np.random.randn(1000)
colors=np.random.rand(1000)
plt.scatter(x,y,s=size,c=colors) 
plt.show()
```

```python
#polar chart, it has two parameters, THETA for the angular cordinates and r for the radial cordinates 
plt.polar([1,2,3,4,5,6,7,8],[4]*8)
plt.show()
```

```python
theta = np.arange(0.,2.,1./180.)*np.pi
plt.polar(theta,np.cos(4*theta),color='green')
plt.polar(3*theta,theta/5,color='red')
plt.polar(theta,[1.4]*len(theta), color='black')
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.polar(theta,color='purple')
plt.show()
```

```python
#using thetagrids and rgrids
theta = np.arange(0.,2.,1./180.)*np.pi
r=np.abs(np.sin(5*theta)-2.*np.cos(theta))
plt.polar(theta,r)
plt.thetagrids(range(45,360,90))
plt.rgrids(np.arange(0.2,3.1,.7),angle=45) #angle is the value to be placed of the numbers showed in the circle
plt.show()
```

```python
#TEXT inside figure,using the text function you can write a string anywhere with respect to x,y axis
#plt.text(x,y,text)
x=np.arange(0,2*np.pi,.01)
y=np.sin(x)
plt.plot(x,y)
plt.text(0.1,-0.04,'sin(0)=0 lies here')
plt.show()
```
