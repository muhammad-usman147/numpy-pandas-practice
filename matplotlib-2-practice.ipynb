```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
```

```python
x=np.array([i for i in range(0,20,2)])
plt.plot(x,x*2)
plt.plot(x,x**0.5)
plt.legend(['multiply * 2','with pow 2'],loc='center left')
plt.grid()
plt.xlim(0,20)
plt.ylim(0,40)
plt.show()
```

```python
mpl.rcParams['figure.figsize']  # showing the 6x4 image displays with an 100 DPI resilts in an 600.400 images pixel
```

```python
#saving the image with different DPI, instead of 100 DPI
x=np.array([i for i in range(0,20,2)])
plt.plot(x,x*2)
plt.plot(x,x**0.5)
plt.legend(['multiply * 2','with pow 2'],loc='center left')
plt.grid()
plt.xlim(0,20)
plt.ylim(0,40)
plt.savefig('graph1.png', dpi=200)
```
